```python
import pandas as pd
import numpy as np

df = pd.read_excel("k5results.xlsx")
df
```

```python
numeric_tf_list = np.array([1 if x == 'T' else 0 for x in df["lst"]])
cumulative_ts = np.cumsum(numeric_tf_list)
```

```python
import matplotlib.pyplot as plt

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(cumulative_ts, marker='o', linestyle='-', label='Cumulative Number of Computational Work')
plt.plot(range(len(cumulative_ts)), range(len(cumulative_ts)), 'k--', label='Full Accuracy')

plt.title('Cumulative Number of Computational Work Found', fontsize=16)
plt.xlabel('Rank', fontsize=16)
plt.ylabel('Counts', fontsize=16)
plt.legend(fontsize=16, loc='lower right')
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.show()
```
