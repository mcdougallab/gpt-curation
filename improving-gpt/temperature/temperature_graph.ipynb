## f1 score trend

```python
# get gpt4 annotations:

#regard unsure as no 

f = pd.read_excel("k5_merged.xlsx")
f = f[:99]
f = f.rename(columns={"gpt3.5_relevant": "gpt3", "3.5 with unsure": "gpt3unsure"})

gpt3unsure = list(f.gpt3unsure.map(dict(yes = "Yes", Yes = "Yes", no="No", No="No", unsure = "No", Unsure = "No"))[:99])

gpt4 = list(f.gpt4_relevant.map(dict(yes = "Yes", Yes = "Yes", no="No", No="No"))[:99])
```

```python
from sklearn.metrics import f1_score

f1_scores = []

for lst in [relevant35_temp_0_attempt_1, relevant35_temp_0_attempt_2,
       relevant35_temp_0_attempt_3, relevant35_temp_05_attempt_1,
       relevant35_temp_05_attempt_2, relevant35_temp_05_attempt_3,
       relevant35_temp_1_attempt_1, relevant35_temp_1_attempt_2,
       relevant35_temp_1_attempt_3, relevant35_temp_15_attempt_1,
       relevant35_temp_15_attempt_2, relevant35_temp_15_attempt_3,
       relevant35_temp_2_attempt_1, relevant35_temp_2_attempt_2,
       relevant35_temp_2_attempt_3]:
    lst = ["No" if item == "Unsure" else item for item in lst]
    f1_scores.append(f1_score(gpt4, lst, average="macro"))

f1_scores  = [round(item, 2) for item in f1_scores]
f1_scores
```
